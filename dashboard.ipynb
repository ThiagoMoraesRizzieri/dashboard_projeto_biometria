```python
# Inicializar o Dash
app = dash.Dash(__name__)

# Layout do app
app.layout = html.Div([
    html.H1(id='titulo-dinamico', style={'textAlign': 'center'}),

    dcc.Dropdown(
        id='cidade-selecao',
        options=[{'label': cidade, 'value': cidade} for cidade in cidades],
        value=cidades[0]  # Seleciona a primeira cidade por padrão
    ),

    dcc.Dropdown(
        id='modelo-selecao',
        options=[
            {'label': 'ARIMA', 'value': 'ARIMA'},
            {'label': 'ARIMAX (SAR)', 'value': 'ARIMAX (SAR)'},
            {'label': 'Dinâmico (SAR)', 'value': 'Dinâmico (SAR)'},
            {'label': 'ARIMAX (GNN)', 'value': 'ARIMAX (GNN)'},
            {'label': 'Dinâmico (GNN)', 'value': 'Dinâmico (GNN)'},
            {'label': 'Floresta Aleatória', 'value': 'Floresta Aleatória'},
            {'label': 'Floresta Aleatória (SAR)', 'value': 'Floresta Aleatória (SAR)'},
            {'label': 'Floresta Aleatória (GNN)', 'value': 'Floresta Aleatória (GNN)'},
            {'label': 'LSTM', 'value': 'LSTM'},
            {'label': 'LSTM (SAR)', 'value': 'LSTM (SAR)'},
            {'label': 'LSTM (GNN)', 'value': 'LSTM (GNN)'},
            {'label': 'LSTM+GNN', 'value': 'LSTM+GNN'}
        ],
        value='ARIMA'  # Modelo padrão ARIMA
    ),

    html.Label("Escolha o horizonte de previsão:"),
    dcc.RadioItems(
        id='horizonte-selecao',
        options=[
            {'label': '7 dias', 'value': 7},
            {'label': '30 dias', 'value': 30},
            {'label': '92 dias', 'value': 92}
        ],
        value=7,  # Valor padrão
        inline=True
    ),

    dcc.Graph(id='grafico-previsao'),
])

# Callback para atualizar o título

@app.callback(
    Output('titulo-dinamico', 'children'),
    [Input('cidade-selecao', 'value'),
     Input('modelo-selecao', 'value')]
)
def atualizar_titulo(cidade_selecionada, modelo_selecionado):
    return f"Previsão do modelo {modelo_selecionado} para a cidade {cidade_selecionada}"

# Callback para atualizar o gráfico
@app.callback(
    Output('grafico-previsao', 'figure'),
    [Input('cidade-selecao', 'value'),
     Input('modelo-selecao', 'value'),
     Input('horizonte-selecao', 'value')]
)
def atualizar_grafico(cidade_selecionada, modelo_selecionado, horizonte):
    indice = df_teste_num_T.index.get_loc(cidade_selecionada)

    fig = go.Figure()

    # Série temporal original
    fig.add_trace(go.Scatter(
        x=df_treino['datahora'],
        y=df_treino_num_T.loc[cidade_selecionada],
        mode='lines',
        name='Série Temporal Original',
        line=dict(color='black')
    ))

    # Série temporal no teste (somente até o horizonte escolhido)
    fig.add_trace(go.Scatter(
        x=df_teste['datahora'][:horizonte],
        y=df_teste_num_T.loc[cidade_selecionada][:horizonte],
        mode='lines',
        name='Série Temporal Teste',
        line=dict(color='gray')
    ))

    if modelo_selecionado == 'ARIMA':
        # Previsão ARIMA
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=arima_predictions[indice][:horizonte],
            mode='lines',
            name='Previsão ARIMA',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=arima_conf_int_95[indice][:horizonte, 0].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=arima_conf_int_95[indice][:horizonte, 1].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% ARIMA'
        ))

        # Intervalo de confiança de 80%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=arima_conf_int_80[indice][:horizonte, 0].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=arima_conf_int_80[indice][:horizonte, 1].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% ARIMA'
        ))

    elif modelo_selecionado == 'ARIMAX (SAR)':
        # Previsão ARIMAX (SAR)
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_sar_predictions[indice][:horizonte],
            mode='lines',
            name='Previsão ARIMAX (SAR)',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_sar_conf_int_95[indice][:horizonte, 0].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_sar_conf_int_95[indice][:horizonte, 1].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% ARIMAX (SAR)'
        ))

        # Intervalo de confiança de 80%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_sar_conf_int_80[indice][:horizonte, 0].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_sar_conf_int_80[indice][:horizonte, 1].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% ARIMAX (SAR)'
        ))

    elif modelo_selecionado == 'ARIMAX (GNN)':
        # Previsão ARIMAX (GNN)
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_gnn_predictions[indice][:horizonte],
            mode='lines',
            name='Previsão ARIMAX (GNN)',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_gnn_conf_int_95[indice][:horizonte, 0].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_gnn_conf_int_95[indice][:horizonte, 1].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% ARIMAX (GNN)'
        ))

        # Intervalo de confiança de 80%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_gnn_conf_int_80[indice][:horizonte, 0].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=sarimax_gnn_conf_int_80[indice][:horizonte, 1].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% ARIMAX (GNN)'
        ))

    elif modelo_selecionado == 'Dinâmico (SAR)':
        # Previsão Dinâmico (SAR)
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_sar_predictions[indice][:horizonte],
            mode='lines',
            name='Previsão Dinâmico (SAR)',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_sar_conf_int_95[indice][0][:horizonte].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_sar_conf_int_95[indice][1][:horizonte].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% Dinâmico (SAR)'
        ))

        # Intervalo de confiança de 80%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_sar_conf_int_80[indice][0][:horizonte].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_sar_conf_int_80[indice][1][:horizonte].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% Dinâmico (SAR)'
        ))

    elif modelo_selecionado == 'Dinâmico (GNN)':
        # Previsão Dinâmico (GNN)
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_gnn_predictions[indice][:horizonte],
            mode='lines',
            name='Previsão Dinâmico (GNN)',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_gnn_conf_int_95[indice][0][:horizonte].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_gnn_conf_int_95[indice][1][:horizonte].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% Dinâmico (GNN)'
        ))

        # Intervalo de confiança de 80%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_gnn_conf_int_80[indice][0][:horizonte].flatten(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=dynamic_gnn_conf_int_80[indice][1][:horizonte].flatten(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% Dinâmico (GNN)'
        ))

    elif modelo_selecionado == 'Floresta Aleatória':
        # Floresta Aleatória
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_predictions[indice][:horizonte],
            mode='lines',
            name='Floresta Aleatória',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_95[indice]["lower_bound"][:horizonte].to_numpy(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_95[indice]["upper_bound"][:horizonte].to_numpy(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% Floresta Aleatória'
        ))

        # Intervalo de confiança de 80%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_80[indice]["lower_bound"][:horizonte].to_numpy(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_80[indice]["upper_bound"][:horizonte].to_numpy(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% Floresta Aleatória'
        ))

    elif modelo_selecionado == 'Floresta Aleatória (SAR)':
        # Floresta Aleatória (SAR)
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_predictions_sar[indice][:horizonte],
            mode='lines',
            name='Floresta Aleatória (SAR)',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_95_sar[indice]["lower_bound"][:horizonte].to_numpy(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_95_sar[indice]["upper_bound"][:horizonte].to_numpy(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% Floresta Aleatória (SAR)'
        ))

        # Intervalo de confiança de 80%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_80_sar[indice]["lower_bound"][:horizonte].to_numpy(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_80_sar[indice]["upper_bound"][:horizonte].to_numpy(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% Floresta Aleatória (SAR)'
        ))

    elif modelo_selecionado == 'Floresta Aleatória (GNN)':
        # Floresta Aleatória (GNN)
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_predictions_gnn[indice][:horizonte],
            mode='lines',
            name='Floresta Aleatória',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_95_gnn[indice]["lower_bound"][:horizonte].to_numpy(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_95_gnn[indice]["upper_bound"][:horizonte].to_numpy(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% Floresta Aleatória (GNN)'
        ))

        # Intervalo de confiança de 80%
        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_80_gnn[indice]["lower_bound"][:horizonte].to_numpy(),
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=rf_conf_int_80_gnn[indice]["upper_bound"][:horizonte].to_numpy(),
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% Floresta Aleatória (GNN)'
        ))

    elif modelo_selecionado == 'LSTM':
        # LSTM
        previsao = lstm_predictions[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=previsao,
            mode='lines',
            name='LSTM',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        lo95 = lower_bound[horizonte][indice].flatten()
        up95 = upper_bound[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=lo95,
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=up95,
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% LSTM'
        ))

        # Intervalo de confiança de 80%
        lo80 = lower_bound_80[horizonte][indice].flatten()
        up80 = upper_bound_80[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=lo80,
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=up80,
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% LSTM'
        ))

    elif modelo_selecionado == 'LSTM (SAR)':
        # LSTM
        previsao = lstm_predictions_sar[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=previsao,
            mode='lines',
            name='LSTM (SAR)',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        lo95 = lower_bound_sar[horizonte][indice].flatten()
        up95 = upper_bound_sar[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=lo95,
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=up95,
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% LSTM (SAR)'
        ))

        # Intervalo de confiança de 80%
        lo80 = lower_bound_80_sar[horizonte][indice].flatten()
        up80 = upper_bound_80_sar[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=lo80,
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=up80,
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% LSTM (SAR)'
        ))

    elif modelo_selecionado == 'LSTM (GNN)':
        # LSTM
        previsao = lstm_predictions_gnn[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=previsao,
            mode='lines',
            name='LSTM (GNN)',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        lo95 = lower_bound_gnn[horizonte][indice].flatten()
        up95 = upper_bound_gnn[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=lo95,
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=up95,
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% LSTM (GNN)'
        ))

        # Intervalo de confiança de 80%
        lo80 = lower_bound_80_gnn[horizonte][indice].flatten()
        up80 = upper_bound_80_gnn[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=lo80,
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=up80,
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% LSTM (GNN)'
        ))

    elif modelo_selecionado == 'LSTM+GNN':
        # Arquitetura integrada LSTM + GNN
        previsao = lstm_predictions_gnn_lstm[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=previsao,
            mode='lines',
            name='Arquitetura integrada LSTM+GNN',
            line=dict(color='blue')
        ))

        # Intervalo de confiança de 95%
        lo95 = lower_bound_gnn_lstm[horizonte][indice].flatten()
        up95 = upper_bound_gnn_lstm[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=lo95,
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=up95,
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.2)',
            name='IC 95% Arquitetura integrada LSTM+GNN'
        ))

        # Intervalo de confiança de 80%
        lo80 = lower_bound_80_gnn_lstm[horizonte][indice].flatten()
        up80 = upper_bound_80_gnn_lstm[horizonte][indice].flatten()

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=lo80,
            fill=None,
            mode='lines',
            line=dict(width=0),
            showlegend=False
        ))

        fig.add_trace(go.Scatter(
            x=df_teste['datahora'][:horizonte],
            y=up80,
            fill='tonexty',
            mode='lines',
            line=dict(width=0),
            fillcolor='rgba(0, 0, 255, 0.4)',
            name='IC 80% Arquitetura integrada LSTM+GNN'
        ))

    fig.update_layout(title=f'Previsão do modelo {modelo_selecionado} em {cidade_selecionada}',
                      xaxis_title='Data',
                      yaxis_title='Valor')

    return fig
```
